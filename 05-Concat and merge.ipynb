```python
Chicago_df=df[df['region']=='Chicago']
```

```python
Chicago_df.set_index('Date', inplace=True)
Chicago_df.head()
Chicago_df.head()
```

```python
concated=pd.concat([Albany_df,Chicago_df],keys=['Albany','Chicago'])
```
